```python
import pandas as pd
import numpy as np
data = pd.read_excel('sz50.xlsx', sheet_name=None, index_col='datetime')
print (data.keys())
```

```python
d1=pd.read_excel('sz50.xlsx', sheet_name='600036.XSHG', index_col='datetime')
```

```python
print(d1.tail())
```

```python
close=d1.close
closea=np.array(close)
print(closea)
```

```python
import talib as ta
ma10=ta.MA(close.values, 10)
print(ma10[-5:])
```

```python
mas10=pd.Series(ta.MA(close.values, 10))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 7))
plt.plot(closea)
plt.plot(mas10)
plt.ylabel('Price')
plt.legend(['600036','Ma10'])
plt.show()
```

```python
plt.figure(figsize=(15, 7))
i=0
for stock in data.keys():
    closear=data[stock]['close'].values
    ROCR100 = ta.ROCR100(closear,5)
    ROCR100Ar=np.array(ROCR100)
    ROCR100S=pd.Series(ROCR100Ar)
    ROCR100S.index=data[stock].index
    plt.plot(ROCR100S)
    i+=1
    if i>=5:
        break
plt.show()
```
